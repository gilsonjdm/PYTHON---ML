# Visualizações em Python
Exemplos de gráficos e simulações usando **NumPy** e **Matplotlib**.
Execute cada célula para ver os resultados.

```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Gráfico de Linha

```python
dias = np.arange(1, 11)
valores = np.random.randint(10, 100, size=10)

plt.figure()
plt.plot(dias, valores, marker='o')
plt.title('Crescimento ao longo dos dias')
plt.xlabel('Dias')
plt.ylabel('Valor')
plt.grid(True)
plt.show()
```

## 2. Gráfico de Barras

```python
linguagens = ['Python', 'JavaScript', 'Java', 'C++', 'Go']
popularidade = [95, 90, 80, 70, 65]

plt.figure()
plt.bar(linguagens, popularidade)
plt.title('Popularidade de Linguagens de Programação')
plt.xlabel('Linguagem')
plt.ylabel('Popularidade')
plt.show()
```

## 3. Histograma

```python
dados = np.random.normal(0, 1, 1000)

plt.figure()
plt.hist(dados, bins=30)
plt.title('Distribuição Normal (Simulação)')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.show()
```

## 4. Caminhada Aleatória (Random Walk)

```python
passos = 100
x = [0]
y = [0]

for i in range(passos):
    x.append(x[-1] + random.choice([-1, 1]))
    y.append(y[-1] + random.choice([-1, 1]))

plt.figure()
plt.plot(x, y)
plt.title('Simulação de Caminhada Aleatória')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

## 5. Função Matemática

```python
x = np.linspace(-10, 10, 200)
y = x**2

plt.figure()
plt.plot(x, y)
plt.title('Gráfico da função y = x²')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
